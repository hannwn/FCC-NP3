```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('medical_examination.csv')
df
```

```python
df['overweight'] = np.where(df['weight'] / (df['height']/ 100)**2 > 25, 1 , 0)
df
```

```python
df['gluc'] = np.where(df['gluc'] == 1, 0, 1)
df
```

```python
df['cholesterol'] = np.where(df['cholesterol'] == 1, 0, 1)
df
```

```python
dfa = pd.melt(df, id_vars = 'cardio', value_vars= ['cholesterol','gluc','smoke','alco','active','overweight']).copy()
dfa
```

```python
dfa = dfa.groupby(['cardio','variable','value']).size().reset_index(name = 'counts')
dfa
```

```python
sns.catplot(
    data = dfa,
    x = 'variable',
    y = 'counts',
    col = 'cardio',
    kind = 'bar',
    hue = 'value'
)
```

```python
dfc = df.copy()
```

```python
dfc = dfc[(dfc['ap_lo'] <= dfc['ap_hi']) &
          (dfc['height'] >= dfc['height'].quantile(0.025)) &
          (dfc['height'] <= dfc['height'].quantile(0.975)) &
          (dfc['weight'] >= dfc['weight'].quantile(0.025)) &
          (dfc['weight'] <= dfc['weight'].quantile(0.975))
          ]
dfc
```

```python
xcor = dfc.corr()
xcor
```

```python
mask = np.triu(np.ones_like(xcor))
sns.heatmap(
    data = xcor,
    mask = mask,
    annot = True,
    fmt = '.1f'
)
plt.show()
```

```python
fig, axes = plt.subplots(figsize = (10,10))
c = np.ma.array(xcor, mask = mask)
axes.set_xticks(ticks = np.arange(0,14), labels = xcor.columns.to_list(), rotation = 'vertical')
axes.set_yticks(ticks = np.arange(0,14), labels = xcor.columns.to_list())
axes.spines['top'].set_visible(False)
axes.spines['left'].set_visible(False)
axes.spines['bottom'].set_visible(False)
axes.spines['right'].set_visible(False)
fig.colorbar(axes.imshow(c), shrink = 0.5)
plt.show()
```
